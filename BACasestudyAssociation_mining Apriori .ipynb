## Apriori Algorithm is used

Apriori algorithm was the first algorithm that was proposed for frequent itemset mining. It was later improved by R Agarwal and R Srikant and came to be known as Apriori. This algorithm uses two steps “join” and “prune” to reduce the search space. It is an iterative approach to discover the most frequent itemsets.

### Association Rules

```python
## Here I have imported essential libraries
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
## import the dataset into Pandas Dataframe. The head() function is used to get the first n rows.
df = pd.read_csv("Online Retail.csv")
df.head()
```

```python
df.shape
```

### Data Preparation

```python
#Data Cleaning
df['Description'] = df['Description'].str.strip() #removes spaces from beginning and end of the description column
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True) #removes duplicate invoice
df['InvoiceNo'] = df['InvoiceNo'].astype('str') #converting invoice number to be type string, needed as the algorithm understand the invoice number as a string data.
df = df[~df['InvoiceNo'].str.contains('C')] #remove the credit transactions,a credit type of invoice is removed with alphabet 'c'
df.head()
```

```python
df['Country'].value_counts() ## the value count or record count for different countries
```

```python
df.shape
```

I am filtering the data on one particular country Separating transactions for Germany

I have  taken Invoice number, description and sum of the quantity.

The entire code will give me a basket of transaction.

```python
mybasket = (df[df['Country'] =="Germany"]
          .groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

### Viewing the transaction basket
In my basket of Germany there are number of products shown as columns 

The value given under a product column is 0 example 10 colour spaceboy pen under invoice number 536527 is 0.0 that means that product was not present in that invoice.

If the number under a product column is 123... that means that product was the part of that particular invoice that many times.

```python
mybasket.head()
```

Converting all positive vaues (all other numbers like 1,2,3 etc) to 1 and others to 0.That is what my algorithm expects as input.

This is what my algorithm of association analysis expects its input data to be in the form of 0 and 1.

```python
def my_encode_units(x):
    if x <= 0:              # all the numbers less than 0 to 0
        return 0 
    if x >= 1:              # all the numbers greater than 1 to 1
        return 1

my_basket_sets = mybasket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE', inplace=True, axis=1) #Removed "postage" as an item , it is not actually required.
```

```python
my_basket_sets
```

### Training Model

Learning purpose: If the number of transactions is 5, your minimum support count would be 5*60/100 = 3.



## Apriori Algorithm 

Generating frequent itemsets.I am running Apriori on my_basket_sets which I have imported from mlxtend.frequent_patterns(in the first cell)

```python
my_frequent_itemsets = apriori(my_basket_sets, min_support=0.07, use_colnames=True)
```

I'm taking my_frequent_itemset as input and out of support lift and confidence I have taken the metric as lift

```python
#generating rules
my_rules = association_rules(my_frequent_itemsets, metric="lift", min_threshold=1) 
```

```python
#viewing top 100 rules
my_rules.head(100)
```

#### Product A 'Round Snack Boxes set  of 4 Woodland' --> Product B 'Plasters in tin Woodland Animals' 
#### Support = 0.074398
#### Confidence = 0.303571
#### Lift = 2.202098

### Making Recommendations 
#### how do we use the above rules to make recomendations?

```python
## Consider the rule table above RULE NUMBER #5 people brought item B 112 times 
my_basket_sets['ROUND SNACK BOXES SET OF4 WOODLAND'].sum()
```

```python
## Consider the rule table above RULE NUMBER #5 people brought item A 47 times 
my_basket_sets['SPACEBOY LUNCH BOX'].sum()
```

#### That means out people who are buying the item 'ROUND SNACK BOXES SET OF4 WOODLAND' 112 times out of which only 47 times the item SPACEBOY LUNCH BOX is being brought.
#### Therefore the rule says that there is a good support, confidence and lift for these combinations so we can recommend SPACEBOY LUNCH BOX to someone who is buying ROUND SNACK BOXES SET OF4 WOODLAND.

```python
#Filtering rules based on condition
my_rules[ (my_rules['lift'] >= 3) &
       (my_rules['confidence'] >= 0.3) ]
```
